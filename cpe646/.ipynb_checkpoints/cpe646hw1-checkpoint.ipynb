# Problem 1

## Problem 1 part 1

```python
import numpy as np
from scipy.stats import norm
import math
import matplotlib.pyplot as plt


x_plot = np.arange(-10,10,0.01)
pxw1 = norm.pdf(x_plot, 0, 1)
pxw2 = norm.pdf(x_plot, 2, 4)
plt.title('Likelihood plot for classes w')
plt.ylabel('Probability')
plt.xlabel('x')
plt.plot(x_plot, pxw1, label='w1')
plt.plot(x_plot, pxw2, label='w2')
plt.legend()
plt.show()

pxw1_5 = norm.pdf(5, 0, 1)
pxw2_5 = norm.pdf(5, 2, 4)
print('Given x = 5, the likelihood for w1 is: %f, and for w2 is: %1.5f' % (pxw1_5, pxw2_5))
```

## Problem 1 part 2

```python
import math

x = 5.0
Pw1 = 0.6
Pw2 = 0.4

px = pxw1 * Pw1 + pxw2 * Pw2
plt.title('Evidence distribution')
plt.xlabel('x')
plt.ylabel('Probability')
plt.plot(x_plot, px)
plt.show()

px5 = pxw1_5 * Pw1 + pxw2_5 * Pw2
print(px5)
```

## Problem 1 part 3

```python
Pw1x = pxw1 * Pw1 / px
Pw2x = pxw2 * Pw2 / px

plt.title('Posterior plot for classes w')
plt.ylabel('Probability')
plt.xlabel('x')
plt.plot(x_plot, Pw1x, label='w1')
plt.plot(x_plot, Pw2x, label='w2')
plt.legend()
plt.show()

Pw1_5x = pxw1_5 * Pw1 / px5
Pw2_5x = pxw2_5 * Pw2 / px5
print('Given x = 5, the posterior probability for w1 is: %f, and for w2 is: %1.5f' % (Pw1_5x, Pw2_5x))
```

## Problem 1 part 4

```python
l = pxw1/pxw2
plt.title('Likelihood ratio')
plt.ylabel('Probability')
plt.xlabel('x')
plt.plot(x_plot, l)
plt.show()

l5 = pxw1_5 / pxw2_5
print(l5)
```

## Problem 1 part 5

lambda is the loss for taking action given a state of nature. It is a discrete number.

lambda(ai|wj)

```python
import numpy as np
lamb = np.array([[0,4],[2,0]])

theta = Pw2 / Pw1
print('The liklihood ratio threshold for a zero-one loss function is: %1.5f' %(theta))
```

## Problem 1 part 6

```python
theta1 = (lamb[0,1] * Pw2) / (lamb[1,0] * Pw1)
print('The liklihood ratio threshold for the risk matrix is: %1.5f' %(theta1))
```

## Problem 1 part 6

```python
ra1x = lamb[0,0] * Pw1x + lamb[0,1] * Pw2x
ra2x = lamb[1,0] * Pw1x + lamb[1,1] * Pw2x
theta_plot = ra2x/ra1x

plt.title('Bayes risk for actions a')
plt.ylabel('Risk')
plt.xlabel('x')
plt.plot(x_plot, ra1x, label='a1')
plt.plot(x_plot, ra2x, label='a2')
plt.legend()
plt.show()

plt.title('Bayes risk ratio')
plt.ylabel('Risk')
plt.xlabel('x')
plt.plot(x_plot, theta_plot)
plt.show()

ra1_5x = lamb[0,0] * Pw1_5x + lamb[0,1] * Pw2_5x
ra2_5x = lamb[1,0] * Pw1_5x + lamb[1,1] * Pw2_5x
print('Given x = 5, the risk of chosing action 1 is: %f, and the risk of chosing action 2 is: %1.5f' % (ra1_5x, ra2_5x))
```

 ## Problem 2 part 2

```python
def k(beta, u1, u2, sigma1, sigma2):
    diff = u2 - u1
    summ = beta * sigma1 + (1 - beta) * sigma2
    pt1 = ((beta * (1 - beta))/2) * diff.T
    pt2 = inv(summ)
    pt3 = np.linalg.norm(summ)
    pt4 = np.linalg.norm(sigma1)**beta * np.linalg.norm(sigma2)**(1 - beta)
    return pt1.dot(pt2).dot(diff) + 0.5 * np.log(pt3/pt4)

bound = np.sqrt(Pw1 * Pw2) * np.exp(-k(0.5, u[:,0], u[:,1], sigma[0,:,:], sigma[1,:,:]))
print(bound)
```
